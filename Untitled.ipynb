```
import pandas as pd
import numpy as np
import pickle as pkl
import tensorflow as tf
import keras.backend as K
from itertools import combinations
from sklearn.model_selection import train_test_split
from keras.backend import l2_normalize
from keras.utils.vis_utils import plot_model
from keras import regularizers
from sklearn.metrics import hamming_loss, accuracy_score 
from sklearn.metrics.pairwise import cosine_distances
```

```
from keras.layers import Input, Dense, concatenate, Dropout
from keras.layers import Lambda
from keras.models import Model
import keras.backend as K
from keras.utils import plot_model
from keras.losses import cosine_similarity
```

```
training_data = pd.read_csv("/content/drive/My Drive/huse/training.csv")
image_embedding = pkl.load(open("/content/drive/My Drive/huse/embedding/image_embedding_4000.pkl",'rb'))
word_embedding = pkl.load(open("/content/drive/My Drive/huse/embedding/word_embedding.pkl",'rb'))
class_embedding_dict = pkl.load(open("/content/drive/My Drive/huse/embedding/class_embedding.pkl",'rb'))
```

```
word_emb = word_embedding[:4001]          # only considering 4000 data 
training_data = training_data[:4001]
```

```
training_data.head()
```

```
print(image_embedding.shape, word_emb.shape , training_data.shape)
```

```
num_class = len(set(training_data["classes"]))
```

Semantic Graph

```
embedding = {}
for i in set(training_data["classes"]):
  i = i.split("<")
  temp = []
  for j in i:
    temp.append(class_embedding_dict[j])
  embedding[" ".join(i)] = np.average(np.array(temp), axis=0)
```

```
training_data["classes"] = list(map(lambda x:str.replace(x,"<"," "), training_data["classes"]))
```

```
labels = list(set(training_data["classes"]))

# Semantic graph 
A_ij = np.zeros((num_class, num_class))
label_combination = list(combinations(range(num_class), 2))
for i in label_combination:
  A_ij[i[0], i[1]] = cosine_distances(embedding[labels[i[0]]], embedding[labels[i[1]]])
```

```
A_ij
```

Labels

```
training_data["labels"] = [labels.index(i) for i in training_data["classes"]]
```

```
training_data.head()
```

```
train_data, test_data= train_test_split(training_data, train_size=0.80)
```

```
train_data.shape
```

```
test_data.shape
```

```
# get train_data
x_train_img = []
x_train_word = []
for i in list(train_data.index):
  x_train_img.append(image_embedding[i])
  x_train_word.append(word_emb[i])
x_train_img = np.array(x_train_img)
x_train_word = np.array(x_train_word)

y_train = np.array(train_data["labels"])


# get test data

x_test_img = []
x_test_word = []
for i in list(test_data.index):
  x_test_img.append(image_embedding[i])
  x_test_word.append(word_emb[i])
x_test_img = np.array(x_test_img)
x_test_word = np.array(x_test_word)

y_test = np.array(test_data["labels"])
```

```
x_train_img.shape, x_train_word.shape, y_train.shape
```

```
x_test_img.shape, x_test_word.shape, y_test.shape
```

```
x_train_img = x_train_img.reshape((x_train_img.shape[0],150528))
x_train_word = x_train_word.reshape((x_train_word.shape[0],1024))

x_test_img = x_test_img.reshape((x_test_img.shape[0],150528))
x_test_word = x_test_word.reshape((x_test_word.shape[0],1024))
```

```
dim = 256
image_dim = (150528,)
text_dim = (1024,)
modelling=True
```

```
input_image = Input(shape=image_dim,  name='input_image')
dense = Dense(512, activation= 'relu', name="dense1")(input_image)
dropout = Dropout(0.15, name="dropout1")(dense)
dense = Dense(512, activation= 'relu', name="dense2")(dropout)
dropout = Dropout(0.15, name="dropout2")(dense)
dense = Dense(512, activation= 'relu', name="dense3")(dropout)                            # image tower
dropout = Dropout(0.15, name="dropout3")(dense)
dense = Dense(512, activation= 'relu', name="dense4")(dropout)
dropout = Dropout(0.15, name="dropout4")(dense)
dense = Dense(512, activation= 'relu', name="dense5")(dropout)
dropout = Dropout(0.15, name="dropout5")(dense)

lmbda = Lambda(lambda x: K.l2_normalize(x, axis=1))(dropout)
image_output = Dense(dim, activation='sigmoid', name="image_output")(lmbda)

input_text = Input(shape=text_dim, name='input_text')
dense = Dense(512, activation= 'relu', name="dense6")(input_text)
dropout = Dropout(0.15, name="dropout6")(dense)
dense = Dense(512, activation= 'relu', name="dense7")(dropout)                           # text tower
dropout = Dropout(0.15, name="dropout7")(dense)

lmbda = Lambda(lambda x: K.l2_normalize(x, axis=1))(dropout)
text_output = Dense(dim, activation='sigmoid', name="text_output")(lmbda)
```

```
x = concatenate([image_output ,text_output]) 
x = Dense(512, activation='relu')(x)
x = Dropout(0.15)(x)
output = Dense(1, activation='sigmoid', name="Output")(x)
```

```
model = Model(inputs=[input_image, input_text],outputs=[output, output, output], name="Huse_model")
plot_model(model, "/content/drive/My Drive/Colab Notebooks/finGenesis/model.png", show_shapes=True)
```

```
def f(y_true, y_pred):
  x_tuples= list(combinations(range(y_true.shape[0]), 2))
  target_tuples = [tuple(y_train[[tuple_]]) for tuple_ in x_tuples]
  l = [A_ij[tuple_] for tuple_ in target_tuples]
  mat = tf.keras.backend.variable(np.array(l))
  l1 = []
  for i in x_tuples:
    temp = [0]*len(y_pred)
    temp[i[0]] = 1
    temp = tf.keras.backend.variable(np.array(temp))

    a = K.sum(tf.keras.layers.Multiply()([y_pred, temp]))
    
    temp = [0]*len(y_pred)
    temp[i[1]] = 1
    temp = tf.keras.backend.variable(np.array(temp))
    
    b = K.sum(tf.keras.layers.Multiply()([y_pred, temp]))
    
    l1.append(A_ij[(a[0],b[0])])
  
  mat1 = tf.keras.backend.variable(np.array(l1))
  return mat, mat1
```

```
def semantic_similarity(y_true, y_pred):
  mat, mat1 = f(y_true, y_pred)
  return tf.keras.losses.MeanSquaredError(mat, mat1).numpy()
```

```
def class_level_loss(y_true, y_pred):
  loss =  tf.keras.backend.binary_crossentropy(
    y_true, y_pred)
  return loss
def loss_function(y_true, y_pred):
  loss1 = class_level_loss(y_true, y_pred) 
  loss2 =  cosine_similarity(
    y_true, y_pred)
  loss3 = semantic_similarity(y_true, y_pred)
  return 0.33*loss1+0.33*loss2+0.5*loss3
```

```
model.compile(optimizer='RMSprop',
              loss= loss_function)
```

```
model.fit({ "input_image": x_train_img, "input_text": x_train_word}, 
          y_train.astype(float),epochs =20, batch_size=32)
```

```
prediction = model.predict([x_test_img, x_test_word])
prediction = np.average(np.array(prediction), axis=0)
```
